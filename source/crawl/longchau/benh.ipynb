```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
from bs4 import BeautifulSoup
from tqdm import tqdm
# Initialize the WebDriver with incognito mode (optional)

def get_main_html(link=None):
# Open the target website
    if link:
        options = Options()
        options.add_argument("--headless")
        driver = webdriver.Chrome(options=options)
        driver.get(link)
        # Click the "load more" button until it's no longer available
        while True:
            try:
                # Wait for the button to be clickable and then click it
                button = WebDriverWait(driver, 30).until(
                    EC.element_to_be_clickable((By.XPATH, "//div[@class='inline-flex justify-center items-center umd:py-2 cursor-pointer']"))
                )
                button.click()
                # Add a short delay after clicking the button
                time.sleep(1)
            except:
                # Break the loop if the button is not found or not clickable
                break

        # Find the target element and get its inner HTML
        html_content = driver.find_element(By.XPATH, "//div[@class='container-lite']").get_attribute('innerHTML')
        domain = driver.find_element(By.XPATH, "//h1[@class='text-display3 umd:text-heading1 text-gray-7 font-bold']").text
        # Close the browser when done
        driver.quit()
        return html_content, domain
```

```python
def parse_html(link, domain, i=0):
    options = Options()
    options.add_argument("--headless")
    driver = webdriver.Chrome(options=options)
    driver.get(link)
    if driver.find_element(By.XPATH, "//button[@class='flex gap-2']"):
        # Wait for the button to be clickable and then click it
        button = WebDriverWait(driver, 30).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@class='flex gap-2']"))
        )
        button.click()
        # Add a short delay after clicking the button
        time.sleep(1)

    page_html = driver.page_source
    driver.quit()
    soup = BeautifulSoup(page_html, "html.parser")
    title = soup.find('h1', class_='text-title3 md:text-display3 font-bold md:mb-4').text.replace('\n', '').replace('\xa0', ' ').replace('\r', '').replace('  ', ' ') if soup.find('h1', class_='text-title3 md:text-display3 font-bold md:mb-4') else None
    html_content = soup.find('div', class_='hidden h-full md:block').find_next_sibling() 
    if not html_content:
        return None
    
    # 3.2. Lấy phần giới thiệu (tìm phần đoạn văn hoặc phần có class đặc biệt)
    intro = None
    intro = html_content.find('div',class_='my-5').text.replace('\n', '').replace('\xa0', ' ').replace('\r', '').replace('  ', ' ') if html_content.find('div',class_='my-5') else None
    # Function to extract content based on the current tag and its siblings
    def extract_section_content(tag):
        content = []
        next_tag = tag.find_next()
        
        # Tiếp tục lặp qua các thẻ kế tiếp cho đến khi gặp một tiêu đề mới (h2, h3, h4...)
        while next_tag and next_tag.name not in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            if next_tag.name == 'p':
                content.append(next_tag.text.strip())
            elif next_tag.name == 'ul':
                list_items = [li.text.strip() for li in next_tag.find_all('li')]
                content.append({'list': list_items})
            elif next_tag.name == 'ol':
                list_items = [li.text.strip() for li in next_tag.find_all('li')]
                content.append({'ordered_list': list_items})
            elif next_tag.name == 'blockquote':
                content.append({'quote': next_tag.text.strip()})
            # Tiếp tục đến thẻ tiếp theo
            next_tag = next_tag.find_next()
        
        return content
    html_content = html_content.find('div', class_='overflow-hidden relative')
    # Trích xuất các mục chính (h1, h2, h3, h4, h5, h6)
    sections = []
    for section in html_content.find_all(['h2', 'h3', 'h4', 'h5', 'h6']):
        section_data = {}
        section_data['heading'] = section.text.strip()
        
        # Lấy nội dung liên quan đến thẻ tiêu đề
        section_data['content'] = extract_section_content(section)

        # Sắp xếp lại các tiêu đề con theo đúng cấp độ h2 -> h3 -> h4...
        if section.name == 'h2':
            sections.append(section_data)
        elif section.name == 'h3':
            if sections:
                if 'subsections' not in sections[-1]:
                    sections[-1]['subsections'] = []
                sections[-1]['subsections'].append(section_data)
        elif section.name == 'h4':
            if sections and 'subsections' in sections[-1]:
                if 'subsections' not in sections[-1]['subsections'][-1]:
                    sections[-1]['subsections'][-1]['subsections'] = []
                sections[-1]['subsections'][-1]['subsections'].append(section_data)
        elif section.name == 'h5':
            if sections and 'subsections' in sections[-1]['subsections'][-1]:
                if 'subsections' not in sections[-1]['subsections'][-1]['subsections']:
                    sections[-1]['subsections'][-1]['subsections'][-1]['subsections'] = []
                sections[-1]['subsections'][-1]['subsections'][-1]['subsections'].append(section_data)
        elif section.name == 'h6':
            if sections and 'subsections' in sections[-1]['subsections'][-1]['subsections'][-1]:
                if 'subsections' not in sections[-1]['subsections'][-1]['subsections'][-1]['subsections']:
                    sections[-1]['subsections'][-1]['subsections'][-1]['subsections'][-1]['subsections'] = []
                sections[-1]['subsections'][-1]['subsections'][-1]['subsections'][-1]['subsections'].append(section_data)
    
    # 4. Chuyển đổi thành JSON và bao gồm link gốc
    data = {
        "title": title,
        "introduction": intro,
        "sections": sections,
        "source_link": link,
        "domain": domain
    }
    
    # 5. Lưu dưới dạng file JSON
    with open(f'benh/article_data{i}.json', 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

    return True
```

```python
def parse_html(link, domain, i=0, driver=None):
    options = Options()
    options.add_argument("--headless")
    # Chỉ mở một trình duyệt duy nhất nếu chưa có
    if driver is None:
        driver = webdriver.Chrome(options=options)
    
    driver.get(link)
    try:
        button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[@class='flex gap-2']"))
        )
        button.click()
        WebDriverWait(driver, 10).until(
            EC.staleness_of(button)  # Đợi cho nút biến mất sau khi click
        )
    except:
        pass  # Nếu không có nút hoặc không bấm được, bỏ qua

    page_html = driver.page_source
    soup = BeautifulSoup(page_html, "html.parser")

    # Xử lý lấy title
    title_elem = soup.find('h1', class_='text-title3 md:text-display3 font-bold md:mb-4')
    title = title_elem.text.strip() if title_elem else None

    html_content = soup.find('div', class_='hidden h-full md:block').find_next_sibling() 
    if not html_content:
        return None

    # Lấy phần giới thiệu
    intro_elem = html_content.find('div', class_='my-5')
    intro = intro_elem.text.strip() if intro_elem else None

    # Function to extract content based on the current tag and its siblings
    def extract_section_content(tag):
        content = []
        next_tag = tag.find_next()

        while next_tag and next_tag.name not in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            if next_tag.name == 'p':
                content.append(next_tag.text.strip())
            elif next_tag.name == 'ul':
                list_items = [li.text.strip() for li in next_tag.find_all('li')]
                content.append({'list': list_items})
            elif next_tag.name == 'ol':
                list_items = [li.text.strip() for li in next_tag.find_all('li')]
                content.append({'ordered_list': list_items})
            elif next_tag.name == 'blockquote':
                content.append({'quote': next_tag.text.strip()})
            next_tag = next_tag.find_next()

        return content

    html_content = html_content.find('div', class_='overflow-hidden relative')
    sections = []
    for section in html_content.find_all(['h2', 'h3', 'h4']):
        section_data = {
            'heading': section.text.strip(),
            'content': extract_section_content(section)
        }
        if section.name == 'h2':
            sections.append(section_data)
        elif section.name == 'h3' and sections:
            if 'subsections' not in sections[-1]:
                sections[-1]['subsections'] = []
            sections[-1]['subsections'].append(section_data)
        elif section.name == 'h4' and sections and 'subsections' in sections[-1]:
            sections[-1]['subsections'][-1].setdefault('subsections', []).append(section_data)

    # Tạo dữ liệu JSON
    data = {
        "title": title,
        "introduction": intro,
        "sections": sections,
        "source_link": link,
        "domain": domain
    }

    # Ghi file JSON
    with open(f'benh/article_data{i}.json', 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

    return True
```

```python
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

# Duyệt qua danh sách article_links và gọi parse_html
for i in range(len(article_links)):
    parse_html(article_links[i][0], article_links[i][1], i, driver)  # Truyền driver vào
    print(f"Processed article {i}")

# Sau khi hoàn tất xử lý, đóng trình duyệt
driver.quit()
```
